```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data.csv')
len(df.columns)
```

```python
df2=df
df2.drop(df2.columns[[0,2,4,6,9,12,14,17,21,22,23,24]], axis = 1,inplace=True)
```

```python
columnsList=df2.columns
#print(columnsList)
dict={columnsList[0]:"platform_look",
     columnsList[1]:"Learn_by_Doing",
     columnsList[2]:"effectivity_learn_by_doing",
     columnsList[3]:"methedology_engagement",
     columnsList[4]:"learning_new",
     columnsList[5]:"widhya_mission",
     columnsList[6]:"mission_effectivity_boolean", 
     columnsList[7]:"hard_level",
     columnsList[8]:"quality_level",
     columnsList[9]:"learning_curve",
     columnsList[10]:"effectivity_microtask",
     columnsList[11]:"platform_satisfaction",
     columnsList[12]:"recommend_us"}
df2.rename(columns=dict,inplace=True)
```

```python
df2.head()
```

```python
%matplotlib inline
print(df2['platform_look'].value_counts())
platform_looks=df2['platform_look'].value_counts()
labels = platform_looks.index
sizes =  platform_looks.values
newlabels=[]
newsizes=[]
count=0
for i in range(len(labels)):
    if labels[i]<6:
        count=count+sizes[i]
    else:
        newlabels.append(str(labels[i]))
        newsizes.append(str(sizes[i]))
newlabels.append("less than 6")
newsizes.append(count)

# print(newlabels)
explode = (0.1,0,0,0,0,0)  
fig1, ax1 = plt.subplots()
ax1.pie(newsizes, explode=explode, labels=newlabels, autopct='%1d%%',
        shadow=True, startangle=170)# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.legend(loc="upper right")
plt.title("Platform rating by students", bbox={'facecolor':'0.8', 'pad':5})
plt.show()


```
